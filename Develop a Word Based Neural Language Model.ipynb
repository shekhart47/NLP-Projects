```python
from keras.preprocessing.text import Tokenizer
from keras.utils import to_categorical
import numpy as np
from keras.utils import to_categorical
from keras.utils.vis_utils import plot_model 
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Embedding
from keras.preprocessing.sequence import pad_sequences
```

```python
data = """ Jack and Jill went up the hill\n
    To fetch a pail of water\n
    Jack fell down and broke his crown\n
    And Jill came tumbling after\n """
```

```python
data
```

```python
def tokenize_data(data):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts([data])
    encoded = tokenizer.texts_to_sequences([data])[0]
    vocab_size= len(tokenizer.word_index)+1
    return tokenizer,encoded,vocab_size
```

```python
tokenizer,encoded,vocab_size = tokenize_data(data)
```

```python
def create_sequences(encoded):
    sequences = []
    for i in range(1, len(encoded)):
        sequence = encoded[i-1:i+1]
        sequences.append(sequence)
    return np.array(sequences)
```

```python
sequences  = create_sequences(encoded)
```

```python
def create_data(sequences,vocab_size):
    sequences = np.array(sequences)
    X = sequences[:,0]
    y = sequences[:,1]
    y = np.array(to_categorical(y, num_classes= vocab_size))
    return X,y
```

```python
X,y  = create_data(sequences,vocab_size)
```

```python
X.shape
```

```python
def make_model(vocab_size):
    model = Sequential()
    model.add(Embedding(input_dim = vocab_size, output_dim = 10, input_length = 1))
    model.add(LSTM(50, activation = 'relu'))
    model.add(Dense(vocab_size, activation = 'softmax'))
    model.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
    return model
```

```python
model = make_model(vocab_size)
```

```python
model.fit(X,y, verbose = 2, epochs = 500)
```

```python
def generate_sequences(seed_text,n_words, tokenizer, model):
    input_text = seed_text
    result = seed_text
    for _ in range(n_words):
        encoded = np.array(tokenizer.texts_to_sequences([input_text])[0])
        y_hat = model.predict_classes(encoded, verbose =0)
        output_word = ''
        for word, index in tokenizer.word_index.items():
            if index == y_hat:
                output_word = word
                break
        result = result + ' ' + output_word
        input_text = output_word
    return result
```

```python
n_words = 6
seed_text = "Jack"
print(generate_sequences(seed_text,n_words, tokenizer, model))
```

# Model 2: Line-by-Line Sequence

```python
def tokenize_data(data):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts([data])
    
    vocab_size= len(tokenizer.word_index)+1
    return tokenizer,vocab_size
```

```python
tokenizer,vocab_size = tokenize_data(data)
```

```python
def create_sequences(encoded, data, tokenizer):
    sequences = []
    for line in data.split('\n'):
        encoded = tokenizer.texts_to_sequences([line])[0]
        for i in range(1, len(encoded)):
            sequence = encoded[:i+1]
            sequences.append(sequence)
    return sequences
```

```python
sequences  = create_sequences(encoded, data, tokenizer)
```

```python
max_len = max([len(item) for item in sequences])
```

```python
max_len
```

```python
def create_data(sequences,vocab_size,max_len):
    #sequences = np.array(sequences)
    sequences = np.array(pad_sequences(sequences, maxlen= max_len, padding = 'pre'))
    X = sequences[:,:-1]
    y = sequences[:,-1]
    y = np.array(to_categorical(y, num_classes= vocab_size))
    return X,y
```

```python
X,y = create_data(sequences,vocab_size,max_len)
```

```python
y.shape
```

```python
def make_model(X, vocab_size):
    model = Sequential()
    model.add(Embedding(input_dim= vocab_size, output_dim= 10, input_length = X.shape[1]))
    model.add(LSTM(50, activation ='relu'))
    model.add(Dense(vocab_size, activation ='softmax'))
    model.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
    return model
```

```python
model = make_model(X,vocab_size)
```

```python
model.fit(X,y, verbose = 2, epochs = 500)
```

```python
def generate_sequences(seed_text, n_words, tokenizer, model, max_len):
    input_text = seed_text
    for _ in range(n_words):
        encoded = tokenizer.texts_to_sequences([input_text])[0]
        encoded = np.array(pad_sequences([encoded], maxlen = max_len))
        y_hat = model.predict_classes(encoded, verbose = 0)
        output_word = ''
        for word, index in tokenizer.word_index.items():
            if index == y_hat:
                output_word = word
                break
        input_text += ' '+ output_word
    return input_text    
```

```python
n_words = 6
seed_text = "Jack"
print(generate_sequences(seed_text,n_words, tokenizer, model ,max_len-1))
n_words = 6
seed_text = "Jill"
print(generate_sequences(seed_text,n_words, tokenizer, model ,max_len-1))
```
